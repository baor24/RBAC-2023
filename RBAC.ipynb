```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt # visualizing data
import seaborn as sns 

explode = (0.1,0,0)  
fig1, ax1 = plt.subplots(figsize=(12,7))
ax1.pie(df['CustomerGender'].value_counts(), explode=explode ,labels=['Male','Female','Unknown'], autopct='%1.1f%%',
        shadow=True, startangle=90)

ax1.axis('equal') 
plt.tight_layout()
plt.legend()
plt.show()
```

```python
province_counts = df['Province'].value_counts()

expected_provinces = ['Hanoi', 'Ho Chi Minh City', 'Southern Provinces', 'Northern Provinces']

# Create a pie chart
fig, ax = plt.subplots(figsize=(12,7))
ax.pie(province_counts, labels=province_counts.index, autopct='%1.1f%%', startangle=90)

# Equal aspect ratio ensures that pie is drawn as a circle
ax.axis('equal')

# Set the title of the pie chart
ax.set_title('Distribution of Provinces')

plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df['TransactionDate'] = pd.to_datetime(df['TransactionDate'], utc=True)

df['Month'] = df['TransactionDate'].dt.to_period('M')

monthly_sales = df.groupby('Month')['SalesAmount'].sum()

monthly_sales.plot(kind='bar', figsize=(10, 6))

plt.title('Total Sales Amount by Month')
plt.xlabel('Month')
plt.ylabel('Total Sales Amount')
plt.xticks(rotation=45)
plt.tight_layout() 
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df['TransactionDate'] = pd.to_datetime(df['TransactionDate'], utc=True)

df['Quarter'] = df['TransactionDate'].dt.to_period('Q')

quarterly_sales = df.groupby('Quarter')['SalesAmount'].sum()

quarterly_sales.plot(kind='bar', figsize=(10, 6))

plt.title('Total Sales Amount by Quarter')
plt.xlabel('Quarter')
plt.ylabel('Total Sales Amount')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

customer_sales = df.groupby('CustomerID')['SalesAmount'].sum().sort_values(ascending=False)

plt.figure(figsize=(12, 6))

top_n = 50
customer_sales.head(top_n).plot(kind='bar')

plt.title(f'Top {top_n} Customers by Total Sales Amount')
plt.xlabel('Customer ID')
plt.ylabel('Total Sales Amount')
plt.tight_layout()  # Adjust layout
plt.show()
```
